## Nature 论文爬取任务 - Task 1
_本 Notebook 用于爬取 Nature 网站关于 LLM（2023～2024）的搜索结果_




### 导入 requests 库

```python
import requests
```

### 步骤 1：构造请求参数


```python
# 手动拼接 URL 以避免参数覆盖问题
# 搜索关键词：llm
# 时间范围：2023-2024
# 页码：第 1 页
search_url = "https://www.nature.com/search?q=llm&date_range=2023-2024&page=1"

# 设置请求头模拟浏览器访问
headers = {
    # 使用常见浏览器 User-Agent
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    # 设置来源页面
    "Referer": "https://www.nature.com/",
    # 指定接受的语言类型
    "Accept-Language": "en-US,en;q=0.9"
}
```

#### 步骤 2：发送 HTTP 请求


```python
# 发送 GET 请求
response = requests.get(search_url, headers=headers)

# 检查响应状态码（200 表示成功）
print(f"响应状态码：{response.status_code}")

# 手动设置编码格式为 utf-8 避免乱码
response.encoding = "utf-8"
```

### 步骤 3：保存结果


```python
# 将响应内容写入文件
with open("page1.txt", "w", encoding="utf-8") as f:
    f.write(response.text)
    print("文件保存成功！文件名为：page1.txt")
```
